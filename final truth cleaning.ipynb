```python
import pandas as pd
df = pd.read_csv("final_cleaned.csv")
df
```

```python
import numpy as np
# array of invalid values

unique_values_age = df['Age_Group'].unique()
unique_values_age
invalid_values_age = unique_values_age[-6:]
invalid_values_age

unique_values_gender = df['Gender'].unique()
unique_values_gender
invalid_values_gender = unique_values_gender[-6:]
invalid_values_gender

unique_values_SES = df['Social_Status'].unique()
unique_values_SES
invalid_values_SES = unique_values_SES[-5:]
invalid_values_SES

invalid_values = np.concatenate((invalid_values_age, invalid_values_SES, invalid_values_gender))
invalid_values
```

```python
df["Age_Group"].replace(invalid_values, "NaN", inplace=True)
df["Gender"].replace(invalid_values, "NaN", inplace=True)
df["Social_Status"].replace(invalid_values, "NaN", inplace=True)
df
```

```python
# checking
df['Age_Group'].unique()
df['Gender'].unique()
df['Social_Status'].unique()
```
