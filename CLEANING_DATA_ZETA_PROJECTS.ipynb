# **App_Categories_Used Cleaned Data**

```python
app_categories_behavioral_unique = behavioral_data['App_Categories_Used'].value_counts()

column_analysis = {
    "App_Categories_Used": app_categories_behavioral_unique,
}
column_analysis
```

```python
app_categories_internet_activity_unique = internet_activity_data['App_Categories_Used'].value_counts()

column_analysis = {
    "App_Categories_Used": app_categories_internet_activity_unique,
}
column_analysis
```

There are some equivalent terms in 'App_Categories_Used', such as:
* Media Sociales → Social Media
* Jeux → Gaming
* Achats → Shopping

So we will mapping of similar terms to unified categories and standardized classification categories:
1. Social Media
2. Gaming
3. Streaming
4. Mixed
5. Shopping




```python
import pandas as pd

# Standardize category names mapping
category_mapping = {
    "Medios sociales": "Social Media",
    "Jeux": "Gaming",
    "Achats": "Shopping"
}

# Function to standardize category names
def standardize_category(category):
    if pd.isna(category) or "This is free text for App_Categories_Used" in str(category):
        return None  # Mark as None for later processing
    return category_mapping.get(category, category)

# Apply standardization to both datasets
behavioral_data["App_Categories_Used"] = behavioral_data["App_Categories_Used"].apply(standardize_category)
internet_activity_data["App_Categories_Used"] = internet_activity_data["App_Categories_Used"].apply(standardize_category)

# Remove duplicates before merging
behavioral_data = behavioral_data.drop_duplicates(subset=['User_ID'], keep='first')
internet_activity_data = internet_activity_data.drop_duplicates(subset=['User_ID'], keep='first')

# Merge datasets on User_ID, preserving both App_Categories_Used columns
merged_df = pd.merge(behavioral_data, internet_activity_data, on="User_ID", how="outer", suffixes=("_behavioral", "_internet"))

# Function to resolve category conflicts
def resolve_category(row):
    behavioral = row['App_Categories_Used_behavioral']
    internet = row['App_Categories_Used_internet']
    if pd.isna(behavioral):
        return internet
    if pd.isna(internet):
        return behavioral
    if behavioral == internet:
        return behavioral
    return "Mixed"

# Apply the resolution function and create a new 'App_Categories_Used' column
merged_df['App_Categories_Used'] = merged_df.apply(resolve_category, axis=1)

final_app_categories_used = merged_df.copy()

# Define the preferred column order
column_order = [
    "User_ID",
    "App_Categories_Used_behavioral",
    "App_Categories_Used_internet",
    "App_Categories_Used",
    "Browsing_Hours_Weekday_behavioral",
    "Browsing_Hours_Weekend_behavioral",
    "Browsing_Hours_Weekday_internet",
    "Browsing_Hours_Weekend_internet",
    "Data_Usage_MB"
]

# Reorder the columns based on the preferred order
final_app_categories_used = final_app_categories_used[column_order]

# Display the final cleaned dataset
display(final_app_categories_used)
```

# **Browsing Hours Cleaned Data**

For **Browsing Hours**, there are some invalid data, such as:  

1. Values greater than 24 (since the maximum for a day is 24 hours).  
2. Data that is not in integer format.  
3. Data with more than two decimal places.  

Similarly, with **App Categories Used**, we will complement each other's data on behavioral and internet activity.

```python
# Make a copy of the dataset to avoid modifying the original DataFrame
final_browsing = final_app_categories_used.copy()

# Define relevant browsing columns
browsing_columns = [
    "Browsing_Hours_Weekday_behavioral", "Browsing_Hours_Weekend_behavioral",
    "Browsing_Hours_Weekday_internet", "Browsing_Hours_Weekend_internet"
]

# Convert values to absolute (remove negative values)
final_browsing[browsing_columns] = final_browsing[browsing_columns].abs()

# Remove invalid values (>24) and non-numeric values
for col in browsing_columns:
    final_browsing[col] = pd.to_numeric(final_browsing[col], errors='coerce')  # Convert non-numeric to NaN
    final_browsing.loc[final_browsing[col] > 24, col] = None  # Mark values >24 as invalid

# Merge behavioral and internet browsing hours (behavioral takes priority)
final_browsing["Browsing_Hours_Weekday"] = final_browsing["Browsing_Hours_Weekday_behavioral"].combine_first(
    final_browsing["Browsing_Hours_Weekday_internet"]
)
final_browsing["Browsing_Hours_Weekend"] = final_browsing["Browsing_Hours_Weekend_behavioral"].combine_first(
    final_browsing["Browsing_Hours_Weekend_internet"]
)

# Ensure values are rounded to two decimal places
final_browsing["Browsing_Hours_Weekday"] = final_browsing["Browsing_Hours_Weekday"].apply(
    lambda x: round(x, 2) if pd.notna(x) else None
)
final_browsing["Browsing_Hours_Weekend"] = final_browsing["Browsing_Hours_Weekend"].apply(
    lambda x: round(x, 2) if pd.notna(x) else None
)

# Function to handle outliers using IQR and addressing NaNs with Median Imputation
def remove_outliers_and_impute_median(col):
    # Calculate quantiles and IQR using only non-NaN values
    Q1 = final_browsing[col].dropna().quantile(0.25)
    Q3 = final_browsing[col].dropna().quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers among non-NaN values
    outliers = ((final_browsing[col] < lower_bound) | (final_browsing[col] > upper_bound)).sum()
    total = final_browsing[col].notna().sum()

    # Replace outliers with median of non-NaN values if outliers are less than 10%
    if outliers / total < 0.05:
        median_value = final_browsing[col].dropna().median()
        final_browsing.loc[
            (final_browsing[col] < lower_bound) | (final_browsing[col] > upper_bound), col
        ] = median_value

    # Impute remaining NaNs with the median of non-NaN values for the current column
    final_browsing[col] = final_browsing[col].fillna(final_browsing[col].dropna().median())

# Apply outlier handling and NaN imputation using median
for col in ["Browsing_Hours_Weekday", "Browsing_Hours_Weekend"]:
    remove_outliers_and_impute_median(col)

# Define the preferred column order
br_column_order = [
    "User_ID",
    "App_Categories_Used_behavioral",
    "App_Categories_Used_internet",
    "App_Categories_Used",
    "Browsing_Hours_Weekday_behavioral",
    "Browsing_Hours_Weekend_behavioral",
    "Browsing_Hours_Weekday_internet",
    "Browsing_Hours_Weekend_internet",
    "Browsing_Hours_Weekday",
    "Browsing_Hours_Weekend",
    "Data_Usage_MB"
]

# Reorder the columns based on the preferred order
final_browsing = final_browsing[br_column_order]

display(final_browsing)
```

```python
# Check for missing values in the merged DataFrame
print(final_browsing.isnull().sum())
# Check the number of rows before and after merging
print("Behavioral Data: ", len(behavioral_data))
print("Internet Activity Data: ", len(internet_activity_data))
print("final app categories used: ", len(final_app_categories_used ))
print("Final Browsing:", len(final_browsing))
```
